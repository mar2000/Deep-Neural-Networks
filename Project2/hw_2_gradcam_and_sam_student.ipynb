```python
class SyntheticData[T](Dataset):
    """A small synthetic segmentation dataset.

    It is a sequence dataset of 5000 tuples (image, class, mask), where:
    - image: before transformation, an RGB PIL Image.
    - class: int 0..4, the label index.
    - mask:  numpy array of dtype=bool, shape (H, W), same size as image.
    """

    CLASSES: Final[tuple[str, ...]] = ("circle", "square", "triangle", "star", "diamond")

    def __init__(
        self,
        path: Path,
        split: Literal["train", "val", "test"],
        transform: Callable[[PIL.Image.Image], T],
    ) -> None:
        self.path = path
        self.transform = transform
        self.split = split

        self.class_to_idx = {class_name: i for i, class_name in enumerate(self.CLASSES)}

        images = sorted((self.path / "images" / self.split).glob("*.png"))
        masks = sorted((self.path / "masks" / self.split).glob("*.png"))
        labels = sorted((self.path / "labels" / self.split).glob("*.txt"))

        assert images, f"No images found in {self.path / 'images' / self.split}"
        assert len(images) == len(masks) == len(labels), (
            "Number of images, masks, and labels must be the same"
        )
        assert [p.stem for p in images] == [p.stem for p in masks] == [p.stem for p in labels], (
            "Image/mask/label filename mismatch."
        )
        self.image_names = [p.stem for p in images]

    def __len__(self) -> int:
        return len(self.image_names)

    def __getitem__(self, idx) -> tuple[T, int, np.ndarray]:
        name = self.image_names[idx]
        img_path = self.path / "images" / self.split / f"{name}.png"
        mask_path = self.path / "masks" / self.split / f"{name}.png"
        label_path = self.path / "labels" / self.split / f"{name}.txt"

        img = PIL.Image.open(img_path).convert("RGB")
        mask = np.array(PIL.Image.open(mask_path), dtype=bool)
        label = self.class_to_idx[label_path.read_text().strip()]

        img_transformed: T = self.transform(img)

        return img_transformed, label, mask


def show_image_row(
    image_dict: dict[str, PIL.Image.Image | np.ndarray | Tensor], size: float = 3.0
) -> None:
    n = len(image_dict)
    _, axs = plt.subplots(1, n, figsize=(size * n, size), constrained_layout=True, squeeze=True)
    if n == 1:
        axs = [axs]
    for ax, (title, img) in zip(axs, image_dict.items(), strict=True):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")
    plt.show()
```

## Training a classifier [do not modify]

```python
def train(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    test_loader: torch.utils.data.DataLoader,
    epochs: int = 5,
    **optimizer_kwargs: Any,
) -> None:
    print(f"ðŸš€ Training CNN for {epochs} epochs...")
    optimizer = optim.AdamW(model.parameters(), **optimizer_kwargs)

    for epoch in range(epochs):
        _train_epoch(model, train_loader, optimizer, desc=f"Epoch {epoch + 1}/{epochs} training  ")
        val_metrics = evaluate(model, val_loader, desc=f"Epoch {epoch + 1}/{epochs} validation")
        print(
            f"Epoch {epoch + 1}/{epochs} â€” "
            + f"val loss: {val_metrics['loss']:.3f}, val acc: {val_metrics['accuracy']:.1%}"
        )

    test_metrics = evaluate(model, test_loader, desc="Test Evaluation")
    print(
        "âœ… Model training complete: "
        + f"Test loss: {test_metrics['loss']:.3f}, test acc: {test_metrics['accuracy']:.1%}"
    )


def _train_epoch(
    model: nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: optim.Optimizer,
    desc: str,
) -> dict[str, float]:
    model.train()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    progress_bar = tqdm(dataloader, desc=desc)
    for imgs, labels, _ in progress_bar:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        _, predicted = torch.max(outputs.data, dim=1)
        total_samples += labels.shape[0]
        total_correct += (predicted == labels).sum().item()

        progress_bar.set_postfix(
            train_loss=f"{total_loss / (total_samples / labels.shape[0]):.3f}",
            train_acc=f"{total_correct / total_samples:.1%}",
        )

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


def evaluate(
    model: nn.Module, dataloader: torch.utils.data.DataLoader, desc: str
) -> dict[str, float]:
    model.eval()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for imgs, labels, _mask in tqdm(dataloader, desc=desc):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = nn.CrossEntropyLoss()(outputs, labels)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, dim=1)
            total_samples += labels.shape[0]
            total_correct += (predicted == labels).sum().item()

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


class DataloaderArgs(TypedDict, total=False):
    batch_size: int
    shuffle: bool
    num_workers: int
    pin_memory: bool
```

## GradCAM implementation (add your code)

```python
class GradCAM:
    """
    Pure PyTorch implementation of Grad-CAM.

    Usage:
        grad_cam = GradCAM(model=model, target_layers=[layer1, layer2])
        grayscale_cam = grad_cam(input_tensor, targets=[class_id])
    """

    def __init__(self, model: nn.Module, target_layers: Iterable[nn.Module]) -> None:
        self.model = model
        self.target_layers = target_layers
        self.activations_and_grads = []

        # Zarejestruj hooki do przechwytywania aktywacji i gradientÃ³w
        self.handles = []
        for layer in target_layers:
            # Hook dla aktywacji (forward pass)
            handle_act = layer.register_forward_hook(
                lambda module, input, output: self.activations_and_grads.append({
                    'activation': output.detach(),
                    'layer': module
                })
            )

            # Hook dla gradientÃ³w (backward pass)
            handle_grad = layer.register_full_backward_hook(
                lambda module, grad_input, grad_output: self.activations_and_grads[-1].update({
                    'gradient': grad_output[0].detach()
                })
            )

            self.handles.extend([handle_act, handle_grad])

    def __del__(self):
        # UsuÅ„ hooki przy usuwaniu obiektu
        for handle in self.handles:
            handle.remove()

    def __call__(self, input_tensor: Tensor, targets: Iterable[int] | None = None) -> np.ndarray:
        """
        Returns: numpy array of shape (B, H, W), values 0..1.
        """
        self.model.eval()
        self.activations_and_grads = []  # WyczyÅ›Ä‡ poprzednie aktywacje

        # Forward pass
        output = self.model(input_tensor)

        # JeÅ›li nie podano targetÃ³w, uÅ¼yj klasy z najwyÅ¼szym prawdopodobieÅ„stwem
        if targets is None:
            targets = [torch.argmax(output[i]).item() for i in range(output.shape[0])]

        # Backward pass
        self.model.zero_grad()
        one_hot_output = torch.zeros_like(output)
        for i, target in enumerate(targets):
            one_hot_output[i, target] = 1

        output.backward(gradient=one_hot_output, retain_graph=True)

        # Oblicz heatmapy dla kaÅ¼dej warstwy
        heatmaps = []
        for item in self.activations_and_grads:
            if 'activation' in item and 'gradient' in item:
                act = item['activation']
                grad = item['gradient']

                # WaÅ¼one aktywacje uÅ›rednione po kanaÅ‚ach
                weights = grad.mean(dim=(2, 3), keepdim=True)  # (B, C, 1, 1)
                weighted_activation = (weights * act).sum(dim=1, keepdim=True)  # (B, 1, H, W)

                # ReLU
                weighted_activation = F.relu(weighted_activation)

                # Normalizacja do zakresu 0-1
                weighted_activation = weighted_activation - weighted_activation.flatten(1).min(1)[0].view(-1, 1, 1, 1)
                weighted_activation = weighted_activation / (weighted_activation.flatten(1).max(1)[0].view(-1, 1, 1, 1) + 1e-8)

                # Resize do oryginalnego rozmiaru obrazu
                cam = F.interpolate(
                    weighted_activation,
                    size=input_tensor.shape[-2:],
                    mode='bilinear',
                    align_corners=False
                ).squeeze(1)  # (B, H, W)

                heatmaps.append(cam.cpu().numpy())

        # UÅ›rednij heatmapy z rÃ³Å¼nych warstw
        if heatmaps:
            final_heatmap = np.mean(heatmaps, axis=0)
        else:
            final_heatmap = np.zeros((input_tensor.shape[0], *input_tensor.shape[-2:]))

        return final_heatmap
```

```python
def heatmap_to_rgb_image(
    heatmap: np.ndarray, min: float | None = None, max: float | None = None
) -> PIL.Image.Image:
    """
    Converts a single-channel heatmap to an RGB pillow image using a colormap.

    Args:
    - heatmap: shape (H, W), will be normalized by mapping min..max to 0..1.
    - min: minimum value for normalization, defaults to heatmap.min().
    - max: maximum value for normalization, defaults to heatmap.max()
    """
    heatmap = heatmap.astype(np.float32)
    if min is None:
        min = heatmap.min()
    if max is None:
        max = heatmap.max()
    heatmap = (heatmap - min) / (max - min + 1e-8)
    heatmap_uint8 = (np.clip(heatmap, 0.0, 1.0) * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    return PIL.Image.fromarray(heatmap_color)
```

```python
class BasicSamPipeline:
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        # The basic pipeline always uses a single center point for each image.
        point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])

        # The basic pipeline always uses a single foreground point, no background points.
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def segment(
        self, images: Tensor, point_coords: np.ndarray, point_labels: np.ndarray
    ) -> Tensor:
        """
        Args:
        - images: normalized images, shape (B, C=3, H, W).
        - point_coords: point coordinates within each image, shape (B, num_points, 2), format (x,y).
            Note the format is not (h,w)=(y,x), but (x,y)!
        - point_labels: point labels, shape (B, num_points), dtype int64.
            Label 1 is foreground (should be in mask), 0 is background (shouldn't be in mask).

        Returns: segmentation masks, shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        assert C == 3, f"Expected images.shape=(B, C=3, H, W), got: {images.shape}"
        num_points = point_coords.shape[1]
        assert point_coords.shape == (B, num_points, 2), f"Expected point_coords.shape=({B=}, num_points, 2), got: {point_coords.shape}"
        assert point_labels.shape == (B, num_points), f"Expected point_labels.shape=({B=}, num_points), got: {point_labels.shape}"

        results = list[Tensor]()
        for image, pt_coords, pt_labels in zip(images, point_coords, point_labels, strict=True):
            sam_predictor.set_image(np.array(inverse_transform(image)))
            masks, scores, _logits = sam_predictor.predict(
                point_coords=pt_coords, point_labels=pt_labels, multimask_output=True
            )
            best_mask = masks[np.argmax(scores)]
            results.append(torch.tensor(best_mask, dtype=torch.bool))
        return torch.stack(results)
```

### Strategy 1: Tylko punkty foreground (bez background)

```python
class ForegroundOnlyPipeline(BasicSamPipeline):
    def __call__(self, images: Tensor) -> tuple[Tensor, np.ndarray]:
        """Zwraca tuple: (maski, punkty)"""
        B, C, H, W = images.shape

        grad_cam = GradCAM(model=model, target_layers=target_layers)
        heatmaps = grad_cam(images)

        point_coords = []
        point_labels = []

        for i in range(B):
            heatmap = heatmaps[i]
            max_pos = np.unravel_index(np.argmax(heatmap), heatmap.shape)
            point_coords.append([(max_pos[1], max_pos[0])])
            point_labels.append([1])

        point_coords = np.array(point_coords)
        point_labels = np.array(point_labels, dtype=np.int64)

        masks = self.segment(images, point_coords, point_labels)
        return masks, point_coords
```

### Strategy 2: Punkty foreground i background

```python
class ForegroundBackgroundPipeline(BasicSamPipeline):
    def __call__(self, images: Tensor) -> tuple[Tensor, np.ndarray]:
        """Zwraca tuple: (maski, punkty)"""
        B, C, H, W = images.shape

        grad_cam = GradCAM(model=model, target_layers=target_layers)
        heatmaps = grad_cam(images)

        point_coords = []
        point_labels = []

        for i in range(B):
            heatmap = heatmaps[i]
            max_pos = np.unravel_index(np.argmax(heatmap), heatmap.shape)

            foreground_points = [(max_pos[1], max_pos[0])]
            background_points = [
                (10, 10),
                (W - 10, 10),
                (10, H - 10),
                (W - 10, H - 10)
            ]

            distances = []
            for bg_point in background_points:
                dist = np.sqrt((bg_point[0] - foreground_points[0][0])**2 +
                              (bg_point[1] - foreground_points[0][1])**2)
                distances.append(dist)

            bg_idx = np.argmax(distances)
            bg_point = background_points[bg_idx]

            points = foreground_points + [bg_point]
            labels = [1, 0]

            point_coords.append(points)
            point_labels.append(labels)

        point_coords = np.array(point_coords)
        point_labels = np.array(point_labels, dtype=np.int64)

        masks = self.segment(images, point_coords, point_labels)
        return masks, point_coords
```

# 3. Funkcje ewaluacyjne

Potrzebujemy funkcji do obliczania metryk:

```python
def compute_iou(pred_mask: np.ndarray, gt_mask: np.ndarray) -> float:
    """Oblicza Intersection over Union dla dwÃ³ch masek binarnych."""
    intersection = np.logical_and(pred_mask, gt_mask).sum()
    union = np.logical_or(pred_mask, gt_mask).sum()
    return intersection / (union + 1e-8)

def compute_center_of_mass(mask: np.ndarray) -> tuple[float, float]:
    """Oblicza Å›rodek masy maski."""
    y_indices, x_indices = np.where(mask)
    if len(y_indices) == 0:
        return (mask.shape[1] // 2, mask.shape[0] // 2)
    center_y = np.mean(y_indices)
    center_x = np.mean(x_indices)
    return (center_x, center_y)  # (x, y) dla zgodnoÅ›ci z SAM

def compute_hit_rate(point_coords: np.ndarray, gt_masks: np.ndarray) -> float:
    """Oblicza, jak czÄ™sto punkty trafiajÄ… w maskÄ™ ground-truth."""
    hits = 0
    total_points = 0

    for i in range(len(point_coords)):
        for point in point_coords[i]:
            x, y = int(point[0]), int(point[1])
            if 0 <= y < gt_masks[i].shape[0] and 0 <= x < gt_masks[i].shape[1]:
                if gt_masks[i][y, x]:
                    hits += 1
            total_points += 1

    return hits / total_points if total_points > 0 else 0

def compute_distance_to_center(point_coords: np.ndarray, gt_masks: np.ndarray) -> float:
    """Oblicza Å›redniÄ… odlegÅ‚oÅ›Ä‡ punktÃ³w od Å›rodka masy maski ground-truth."""
    distances = []

    for i in range(len(point_coords)):
        center_x, center_y = compute_center_of_mass(gt_masks[i])

        for point in point_coords[i]:
            x, y = point
            dist = np.sqrt((x - center_x)**2 + (y - center_y)**2)
            distances.append(dist)

    return np.mean(distances) if distances else 0
```

# 4. Ewaluacja caÅ‚ego pipeline'u

Dodajmy funkcjÄ™ do ewaluacji:

```python
def evaluate_pipeline(pipeline, dataloader, name="Pipeline"):
    """Ewaluuje pipeline na caÅ‚ym datasetcie."""
    all_ious = []

    for images, labels, gt_masks in tqdm(dataloader, desc=f"Evaluating {name}"):
        # Uruchom pipeline
        output = pipeline(images.to(device))

        # SprawdÅº typ outputu
        if isinstance(output, tuple):
            pred_masks = output[0]  # weÅº maski z tuple
        else:
            pred_masks = output  # bezpoÅ›rednio maski

        pred_masks_np = pred_masks.cpu().numpy()
        gt_masks_np = gt_masks.numpy()

        # Oblicz IoU dla kaÅ¼dego obrazu
        for i in range(len(pred_masks_np)):
            iou = compute_iou(pred_masks_np[i], gt_masks_np[i])
            all_ious.append(iou)

    # Podsumowanie
    mean_iou = np.mean(all_ious)
    print(f"{name} Results:")
    print(f"  Mean IoU: {mean_iou:.3f} ({mean_iou*100:.1f}%)")
    print(f"  IoU std: {np.std(all_ious):.3f}")
    print(f"  Min IoU: {np.min(all_ious):.3f}")
    print(f"  Max IoU: {np.max(all_ious):.3f}")

    return {
        'mean_iou': mean_iou,
        'all_ious': all_ious
    }
```

```python
def evaluate_points(pipeline, dataloader, name="Pipeline"):
    """Oblicza hit rate i distance dla generowanych punktÃ³w."""
    all_hit_rates = []
    all_distances = []

    for images, labels, gt_masks in tqdm(dataloader, desc=f"Evaluating points for {name}"):
        # Uruchom pipeline - zwraca teraz (maski, punkty)
        pred_masks, point_coords = pipeline(images.to(device))
        gt_masks_np = gt_masks.numpy()

        # Dla kaÅ¼dego obrazu w batchu
        for i in range(len(point_coords)):
            gt_mask = gt_masks_np[i]

            # Hit rate dla tego obrazu
            hits = 0
            total_points = 0

            for point in point_coords[i]:
                x, y = int(point[0]), int(point[1])
                # SprawdÅº czy punkt jest w granicach obrazu
                if 0 <= y < gt_mask.shape[0] and 0 <= x < gt_mask.shape[1]:
                    if gt_mask[y, x]:  # Punkt trafiÅ‚ w maskÄ™
                        hits += 1
                total_points += 1

            hit_rate = hits / total_points if total_points > 0 else 0
            all_hit_rates.append(hit_rate)

            # Distance dla tego obrazu (Å›rodek masy maski)
            y_indices, x_indices = np.where(gt_mask)
            if len(y_indices) > 0:
                center_y = np.mean(y_indices)
                center_x = np.mean(x_indices)

                # Oblicz Å›redniÄ… odlegÅ‚oÅ›Ä‡ punktÃ³w od Å›rodka
                distances = []
                for point in point_coords[i]:
                    x, y = point
                    dist = np.sqrt((x - center_x)**2 + (y - center_y)**2)
                    distances.append(dist)

                avg_distance = np.mean(distances) if distances else 0
                all_distances.append(avg_distance)

    # Oblicz Å›rednie
    mean_hit_rate = np.mean(all_hit_rates) if all_hit_rates else 0
    mean_distance = np.mean(all_distances) if all_distances else 0

    print(f"{name} Point Metrics:")
    print(f"  Mean Hit Rate: {mean_hit_rate:.3f} ({mean_hit_rate*100:.1f}%)")
    print(f"  Mean Distance from Center: {mean_distance:.2f} pixels")
    print(f"  Sample size: {len(all_hit_rates)} images")

    return {
        'mean_hit_rate': mean_hit_rate,
        'mean_distance': mean_distance,
        'all_hit_rates': all_hit_rates,
        'all_distances': all_distances
    }
```
